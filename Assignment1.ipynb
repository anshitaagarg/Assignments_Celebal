```python
import pandas as pd

df = pd.read_csv('Combined_dataset.csv')
print(df.shape)
print(df.columns.tolist())
```

```python
df.info()
```

```python
#Explore Data (head/tail/shape/columns/dtypes)
print("First 5 rows:")
print(df.head())

print("\nLast 5 rows:")
print(df.tail())

print(f"\nShape: {df.shape}")
print(f"\nColumns: {df.columns.tolist()}")
print(f"\nData Types:\n{df.dtypes}")
```

```python
print("Missing values per column:")
print(df.isnull().sum())

# Fill missing 'discount' with 0 (no discount)
df['discount'] = df['discount'].fillna(0)

# Fill missing 'what_customers_said' with 'No reviews'
df['what_customers_said'] = df['what_customers_said'].fillna('No reviews')

# Drop rows where seller_name is missing
df = df.dropna(subset=['seller_name'])

# Fill remaining missing with 'Unknown'
df['videos'] = df['videos'].fillna('No video')
df['variations'] = df['variations'].fillna('No variations')
df['seller_information'] = df['seller_information'].fillna('Unknown')

print("\nMissing values after handling:")
print(df.isnull().sum())
```

```python
# Filter: products with rating >= 4.0
high_rated = df[df['rating'] >= 4.0]
print(f"High rated products (>=4.0): {len(high_rated)}")

# Select specific columns
selected = df[['product_id', 'title', 'initial_price', 'discount', 'rating', 'category']]
print(selected.head())
```

```python
print(f"Before removing duplicates: {df.shape}")
df = df.drop_duplicates()
print(f"After removing duplicates: {df.shape}")
```

```python
# Clean final_price (it's object type, need to convert)
df['final_price'] = df['final_price'].astype(str).str.replace('[^0-9.]', '', regex=True)
df['final_price'] = pd.to_numeric(df['final_price'], errors='coerce').fillna(0)

# total_amount = initial_price * ratings_count (simulating price * quantity)
df['total_amount'] = df['initial_price'] * df['ratings_count']
print(df[['initial_price', 'ratings_count', 'total_amount']].head())
```

```python
df.to_csv('cleaned_shopping_data.csv', index=False)
print("Cleaned dataset saved!")
```

```python
print("========== DATASET SUMMARY ==========")
print(f"Total records after cleaning : {len(df)}")
print(f"Total columns                : {len(df.columns)}")
print(f"\nMissing values remaining:\n{df.isnull().sum()}")
print(f"\nNumerical Stats:")
print(df[['rating', 'initial_price', 'discount', 'total_amount']].describe())
print(f"\nTop 5 categories:")
print(df['category'].value_counts().head())
```
